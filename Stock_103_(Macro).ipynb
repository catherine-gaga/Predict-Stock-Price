```python
# Import packages
import pandas as pd
import numpy as np
import requests
import yahoo_fin.stock_info as si
```

```python
# Get ticker list
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
html_data = requests.get(url).text
df_list = pd.read_html(html_data)
df = df_list[0]  # The first table contains the list of companies
tickers = df['Symbol'].tolist() #list(df['Symbol']) choose the symbol and change to list version
tickers = [s.replace('.', '-') for s in tickers]
print(tickers)
```

```python
# Create function that give final output
def get_adj_per_table(ticker):
  try:
    info = si.get_analysts_info(ticker)
    price = si.get_data(ticker)
    current_price = price['close'].iloc[-1]
    eps_est1 = info['Earnings Estimate'].iloc[1,3]
    eps_est2 = info['Earnings Estimate'].iloc[1,4]
    d = eps_est2 - eps_est1
    a = d
    b = -d+2*eps_est1
    c = -2*current_price
    dis = (b**2) - (4*d*c)
    n = (-b + np.sqrt(dis))/(2 * a)
    df = pd.DataFrame({
      'ticker': ticker,
      'current_price': current_price,
      'eps_est1': eps_est1,
      'eps_est2': eps_est2,
      'Adj_PER': n
                },index=[0])
    print(df)
    return df
      except:
    pass  # Do nothing and continue execution
```

```python
# Summarise output into a table
df = pd.DataFrame() #create the empty table
for ticker in tickers:
  df1= get_adj_per_table(ticker)
  df = pd.concat([df, df1], ignore_index=True)
```

```python
# Clean table
df = df.sort_values(by=['Adj_PER'])#Adj_PER in order
df = df[df['Adj_PER'] > 0]#without negative AdjPER
```
