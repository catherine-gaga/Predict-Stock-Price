```python
import yfinance as yf
from prophet import Prophet
import pandas as pd

# Specify the stock symbol and the period for which you want the historical data
stock_symbol = 'JPYCNY=X'  # Example: Apple Inc.
period = '10y'  # 5 years of historical data

# Fetch historical stock prices
stock_data = yf.download(stock_symbol, period=period)

# Reset the index to make Date a column
stock_data.reset_index(inplace=True) # Same as stock_data = stock_data.reset_index()

# Prepare the data for Prophet
# Prophet requires the column names to be 'ds' (Date) and 'y' (value)
prophet_data = stock_data[['Date', 'Close']].rename(columns={'Date': 'ds', 'Close': 'y'})

# Initialize a Prophet model
model = Prophet(weekly_seasonality=False,yearly_seasonality=False)

# Fit the model with your data
model.fit(prophet_data)

# Create a dataframe for future dates to predict (for example, next 365 days)
future = model.make_future_dataframe(periods=365)

# Use the model to make predictions
forecast = model.predict(future)

# Plot the forecast
fig = model.plot(forecast)
fig.show()

# Plot the forecast components
fig2 = model.plot_components(forecast)
fig2.show()
```

```python
# Change column name before Inner Join
stock_data_EWJ.rename(columns={'Close': 'Close_EWJ'}, inplace=True)# Change the index into column
stock_data_JPY.rename(columns={'Close': 'Close_JPY'}, inplace=True)
```

```python
# Performing an inner join on 'Date'
df = pd.merge(stock_data_EWJ, stock_data_JPY, on='Date', how='inner')
```

```python
# Get Correlation coefficient of Close_EWJ and Close_TNX
correlation = df['Close_EWJ'].corr(df['Close_JPY'])
```

```python
correlation
```

```python
prophet_data = df.rename(columns={'Date': 'ds', 'Close_EWJ': 'y'})
```

```python
prophet_data
```

```python
# Add regressor
# Initialize Prophet model and add the additional regressor
model = Prophet(weekly_seasonality=False,yearly_seasonality=False,daily_seasonality=False)
model.add_regressor('Close_JPY')

# Fit the model with the data
model.fit(prophet_data)

# Make a future dataframe for forecasting
future = model.make_future_dataframe(periods=0)
future = future.merge(prophet_data[['ds', 'Close_JPY']], on='ds', how='left')
future['Close_JPY'] = future['Close_JPY'].ffill()

# Forecast
forecast = model.predict(future)

# Displaying the forecasted stock price
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```

```python
# Plot the forecast
fig = model.plot(forecast)
fig.show()

# Plot the forecast components
fig2 = model.plot_components(forecast)
fig2.show()
```
